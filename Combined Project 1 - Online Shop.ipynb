### Поиск пропущенных значений

```python
# смотрим кол-во пропущенных значений по всем столбцам
print(df.isna().sum())
# выводим данные, где пропущены значения рейтинга ESRB
df[df['rating'].isnull()]
```

Очень много пропущенных данных в столбцах с рейтингом. Большая часть из них связана скорее всего с тем, что ESRB является канадско-американской системой рейтинга игр. В то время как в Японии и и Европе существуют свои рейтинговые системы - CERO и PEGI соотвественно. Выведя на экран те строки, где рейтинг от ESRB нулевой - мы видим, что большая часть из этих игр выпущена не в Америке или Канаде, а значит - и рейтинга для них нет. Заполнить пропущенные данные чем-нибудь валидным не представляется возможным. Оставим все значения как NaN, если мы заменим их на какую-нибудь "заглушку" в виде "unknown", то это сильно усложнит нам процесс анализа данных.

```python
# выведем две игры, без названия
df[df['name'].isnull()]
# удаляем строки без названия игры
df = df[~(df['name'].isnull())]
```

В целом можем не удалять эти две строки - они ни на что не повлияют, но тем не менее. Посмотрим сколько годов со значением NaN.

```python
display(df[df['year_of_release'].isnull()].count())
df = df[~(df['year_of_release'].isnull())]
```

Удалим 269 записей, где year_of_release NaN - впоследствии это будет мешать, а 269 строк особо ни на что не повлияют

### Операции с типами данных

Столбец year_of release следует перевести из типа данных float в int. Год не может быть дробным.

```python
# заменяем значения tbd в столбце user_score на NaN
df['user_score'] = df['user_score'].replace('tbd', np.NaN)
# меняем тип данных
df['user_score'] = df['user_score'].astype(float, errors = 'ignore')
df['year_of_release'] = df['year_of_release'].astype(int, errors = 'ignore')
df.info()
```

### Как менялись продажи по платформам. Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?

```python
# строим график с суммарными продажами по каждой платформе
df.groupby('platform')['total_sales'].sum().sort_values().plot(kind = 'bar', grid=True, figsize=(10,10), title = 'Продажи игр на каждой платформе за все года')
plt.ylabel('Число продаж игр, млн')
plt.xlabel('Платформа')
# соединяем 5 платформ с наибольшими суммарными продажами в один датафрейм
df_five = df[df['platform'].isin(['X360', 'PS3', 'PS2', 'Wii', 'DS'])]
```

Как мы видим - X360, PS3, PS2, Wii, DS имеют наибольшие объемы продаж игр за все годы.

```python
# строим распределение трех платформ с наибольшими суммарными продажами по годам
df_five.pivot_table(index = 'year_of_release', columns = 'platform', values = 'total_sales', aggfunc = 'sum').plot(xlim = (1997, 2017),kind = 'line', grid = True, legend = True, figsize = (15,10), title = 'Продажи игр на 5 платформах за все года')
plt.ylabel('Число продаж игр, млн')
plt.xlabel('Годы')
plt.show()
```

В среднем, после того как платформа выходит на рынок - она остается востребованной в течении 4 лет. Далее объем покупок снижается до 0 в течении последующих 6 лет. Значит, что платформы живут по 10 лет. Средним сроком жизни платформ исходя из вашего кода ниже можно назвать срок в 7 лет.

```python
# Комментарий ревьюера
temp = df.copy() 
time_life = temp.pivot_table(index='platform', values='year_of_release', aggfunc=['min', 'max']).reset_index()
time_life['life'] = time_life['max'] - time_life['min'] + 1 # в срок жизни платформы, должны входить границы диапазона, 
                                                            # поэтому +1
time_life['life'].median()
```

## Проверка гипотез
В этой главе нам предстоит проверить следующие гипотезы:
* cредние пользовательские рейтинги платформ Xbox One и PC одинаковые
* cредние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
* H0 - cредние пользовательские рейтинги платформ Xbox One и PC одинаковые
* H1 - cредние пользовательские рейтинги платформ Xbox One и PC различаются

```python
# пользовательские рейтинги XOne
xone_rat = df_cut[df_cut['platform'] == 'XOne']['user_score']
# пользовательские рейтинги PC
pc_rat = df_cut[df_cut['platform'] == 'PC']['user_score']
# уровень статистической значимости
alpha = 0.05

results = st.ttest_ind(xone_rat, pc_rat, alternative='two-sided', nan_policy='omit')

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
    print()
print()
print('Среднее значение пользовательского рейтинга XOne')
print()
print(xone_rat.mean())
print()
print('Среднее значение пользовательского рейтинга PC')
pc_rat.mean()
```

Значение p-value около 10%. H0 не опровергнута. С вероятностью в 10% можно получить различность рейтингов XOne и PC.

### Cредние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные
* H0 - cредние пользовательские рейтинги жанров Action и Sports одинаковые
* H1 - cредние пользовательские рейтинги жанров Action и Sports различаются

```python
# пользовательские рейтинги Action
action_rat = df_cut[df_cut['genre'] == 'Action']['user_score']
# пользовательские рейтинги Sports
sport_rat = df_cut[df_cut['genre'] == 'Sports']['user_score']
# уровень статистической значимости
alpha = 0.05

results = st.ttest_ind(action_rat, sport_rat, alternative='two-sided', nan_policy='omit')

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
    print()
print()
print('Среднее значение пользовательского рейтинга XOne')
print()
print(action_rat.mean())
print()
print('Среднее значение пользовательского рейтинга PC')
sport_rat.mean()
```

Получив данное p-value, мы отвергли H0. Вероятности получить одинаковые средние рейнтинги по Action и Sports почти нет.

### Вывод по главе

* Так как нам нужно проверить гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые, то для того, чтобы это опровергнуть нужна альтернативная гипотеза - говорящая, что они разные. Тогда "средние пользовательские рейтинги платформ Xbox One и PC одинаковые" - станет H0.
* Так как нам нужно проверить гипотезу, что средние пользовательские рейтинги жанров Action и Sports разные, то для того, чтобы это проверить нужна нулевая гипотеза - говорящая, что они одинаковые. Тогда "средние пользовательские рейтинги жанров Action и Sports разные" - станет H1.
* был использован t-тест потому, что это параметрический тест, который показывает, насколько значительны различия между группами. Другими словами, он позволяет узнать, могли ли эти различия быть получены случайно.
